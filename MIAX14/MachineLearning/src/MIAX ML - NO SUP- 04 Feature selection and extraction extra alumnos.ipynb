## Carga y Preparación

Primero, utilizaremos el "German Credit Risk Dataset", un dataset con múltiples variables financieras (edad, ingresos, duración del préstamo, ahorros, etc.). El objetivo es predecir la solvencia (alta o baja) de cada cliente.

```python
data = fetch_openml(name='credit-g', version=1, as_frame=True)
X = data.data
y = data.target
```

Por simplificar, vamos a quedarnos solo con las variables numéricas.

```python
X = X.select_dtypes(include=[np.number])
y = y.map({'good': 1, 'bad': 0})
```

```python
print(X.shape, y.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```
